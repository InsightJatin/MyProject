```python
features = a.drop(['quality','Winequality'],axis=1)
target = a['Winequality']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(features,target,test_size=0.2,random_state=0)
xtrain.shape,xtest.shape
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean')
xtrain = imputer.fit_transform(xtrain)
xtest = imputer.fit_transform(xtest)
xtrain.shape,xtest.shape
```

```python
from sklearn import metrics
```

```python
models = [LogisticRegression(),XGBClassifier(),SVC(kernel='rbf')]
```

```python
for i in range(3):
    models[i].fit(xtrain,ytrain)
    print(f'{models[i].__class__.__name__}:')

    print('Training accuracy: ',metrics.roc_auc_score(ytrain,models[i].predict(xtrain)))
    print('Validiation accuracy: ',metrics.roc_auc_score(ytest,models[i].predict(xtest)))
    print()
```

```python
def preprocess_new_data(df):
    df = df.copy()
    df['type'] = df['type'].replace({'white':0,'red':1})

    if 'total sulfur dioxide' in df.columns:
        df = df.drop('total sulfur dioxide',axis=1)

    drop_cols = [c for c in['quality','winequality']if c in df.columns]
    df = df.drop(drop_cols,axis=1)

    x = imputer_save.transform(df)
    x = scaler_save.transform(x)
    return x
```

```python
new_data = pd.read_csv("new_wine_samples.csv")
```

```python
new_data.head()
```

```python
x_new = preprocess_new_data(new_data)
```

```python
pred = models[2].predict(x_new)
```

```python
new_data['predicted_winequality'] = pred
```

```python
new_data['quality_label'] = new_data['predicted_winequality'].map({1:'Good(>5)',0:'Bad(<5)'})
```
